```python
df_aus = pd.read_csv("df_aus.csv", na_values = ["-"])
df_q2 = pd.read_csv("df_q2.csv", na_values = ["-"])
df_aus
```

```python
df_q2
```

```python
x = df_aus["date"]
y3 = df_aus["new_tests"]
y2 = df_aus["new_vaccinations"]
y1 = df_aus["new_cases"]
y = df_aus['Price']

fig, ax1 = plt.subplots()
ax1.set_xlabel('Month-Year', fontsize = 15)
ax1.plot(x, y3, linewidth=1, color="peachpuff", label = "new_tests")
ax1.plot(x, y1, linewidth=1, color="red", label = "new_cases")
ax1.plot(x, y2, linewidth=1, color="green", label = "new_vaccinations")
ax1.set_ylabel('total number of subjects', color='b', fontsize = 13)
ax1.legend(loc=6)
ax1.set_xticks(())

ax2 = ax1.twinx()
ax2.plot(x, y, color='paleturquoise', linewidth=1, label = "price")
ax2.set_ylabel('Price', fontsize = 12)
ax2.set_xticks(())
ax2.set_title("Australia covid data")

plt.legend(loc='upper left')
fig1 = plt.gcf()
fig1.set_size_inches(9, 5)
fig1.savefig('multi _variable_price_date_aus.png', format='png')
```

```python
df_q2_Date = df_q2.groupby("Date").mean()
df_q2_Date['new_tests']
```

```python
x = df_q2_Date["Date"]
y3 = df_q2_Date["new_tests"]
y2 = df_q2_Date["new_vaccinations"]
y1 = df_q2_Date["new_cases"]
y = df_q2_Date['Price']

fig, ax1 = plt.subplots()
ax1.set_xlabel('Month-Year', fontsize = 15)
ax1.plot(x, y3, linewidth=1, color="peachpuff", label = "new_tests")
ax1.plot(x, y1, linewidth=1, color="red", label = "new_cases")
ax1.plot(x, y2, linewidth=1, color="green", label = "new_vaccinations")
ax1.set_ylabel('total number of subjects', color='b', fontsize = 13)
ax1.legend(loc=6)
ax1.set_xticks(())

ax2 = ax1.twinx()
ax2.plot(x, y, color='paleturquoise', linewidth=1, label = "price")
ax2.set_ylabel('Price', fontsize = 12)
ax2.set_xticks(())
ax2.set_title("covid data")

plt.legend(loc='upper left')
fig1 = plt.gcf()
fig1.set_size_inches(9, 5)
fig1.savefig('multi _variable_price_date_all.png', format='png')
```
